```python
#exemplo 4: utilizando modulos
def opcao_1():
  return("opção 1 selecionada")
def opcao_2():
  return("opção 2 selecionada")
def opcao_3():
  return("opção 3 selecionada")
def default():
  return("ta errado chefe")
def switch(opcao):
  opcoes = { # Corrected: Changed 'opcao' to 'opcoes' and assigned a dictionary
      1: opcao_1,
      2: opcao_2,
      3: opcao_3,
  }
  func = opcoes.get(opcao,default) # Corrected: Changed 'opcoes' to 'opcoes'
  return func() # Corrected: Changed 'selecao' to 'func' and called the function
```

```python
entrada = int(input("manda uma opcao de 1 a 3 pai:"))
resultado = switch(entrada)
print(resultado)
```

```python
# Tabuleiro de Xadrez
tamanho = 8

# Linhas X Colunas
# Loop de linhas
for linha in range(tamanho):
  # Loop de colunas
  for coluna in range(tamanho):
    if(linha + coluna) % 2 == 0:
      print("▢",end="") # Casa Branca
    else:
      print("▣",end="") # Casa Preta
  print() # Nova linha
```

```python
# Aparelhos: (nome, watts, horas por dia)
aparelhos = [
    ("Geladeira", 150, 24),
    ("Micro-ondas", 1200, 0.5),
    ("TV", 100, 5),
    ("Ar-condicionado", 2000, 8),
    ("Máquina de lavar", 500, 1)
]

dias = 30
limite = 250

# Calcula consumo de cada um
consumos = []
for nome, watts, horas in aparelhos:
    consumo = (watts * horas * dias) / 1000
    consumos.append((nome, consumo))

# Soma tudo
total = 0
for nome, consumo in consumos:
    total += consumo

# Média
media = total / len(consumos)

# Ver quem gasta mais que a média
acima_media = []
for nome, consumo in consumos:
    if consumo > media:
        acima_media.append(nome)

# Mostra os resultados
for nome, consumo in consumos:
    print(nome, "-", round(consumo, 2), "kWh")

print("\nTotal:", round(total, 2), "kWh")
print("Acima da média:", acima_media)

if total > limite:
    print(" Passou do limite de 250 kWh! Melhora isso cara")
else:
    print("✅ Dentro do limite. Ta suave")
```

```python
# Cada tupla representa: (nome do investimento, capital inicial, taxa mensal em %)
investimentos = [
    ("CDB Brabo", 800, 1.5),
    ("Tesouro Direto", 1000, 1.0),
    ("Fundos de Cria", 1500, 2.2),
    ("Poupança", 1000, 0.5)
]

resultados = []            # Guarda os nomes e montantes
rendimentos_altos = []     # Guarda os que passaram de R$1000
dobrou = False             # Se algum dobrou o valor
somas_percentuais = 0      # Soma dos crescimentos percentuais

for nome, capital, taxa in investimentos:
    # Fórmula dos juros compostos: M = C * (1 + i)^t
    montante = capital * (1 + taxa / 100) ** 12

    crescimento_percentual = ((montante - capital) / capital) * 100
    somas_percentuais += crescimento_percentual

    resultados.append((nome, montante))

    if montante > 1000:
        rendimentos_altos.append(nome)

    if montante >= 2 * capital:
        dobrou = True

# Print dos resultados sem arredondar
print("Resultado dos investimentos após 12 meses:\n")
for nome, valor in resultados:
    print(f"{nome}: R$ {valor}")

print("\nInvestimentos com rendimento acima de R$1000:")
print(rendimentos_altos)

print("\nAlgum investimento dobrou de valor?")
print("Sim, chefia!" if dobrou else "Não dobrou não, meu cria...")

media_crescimento = somas_percentuais / len(investimentos)
print(f"\nMédia de crescimento percentual: {media_crescimento}%")
```
